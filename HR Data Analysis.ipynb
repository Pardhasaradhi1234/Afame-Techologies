## Data Preprocessing and Cleaning

```python
df.isnull().sum()
```

- NO NULL values so
### Categorical Data Handling

```python
df.Attrition = df.Attrition.map({'Yes':1,'No':0})
```

```python
df.BusinessTravel = df.BusinessTravel.map({'TravelFrequently':1,'TravelRarely':2,'Non-Travel':0})
```

```python
df.Department = df.Department.map({'ResearchDevelopment':2,'Sales':1,'HumanResources':0})
```

```python
df.EducationField = df.EducationField.map({'LifeSicences':5,'Medical':4,'Marketing':3,'TechnicalDegree':2,'Other':1,'HumanResources':0})
```

```python
df.Gender = pd.get_dummies(df.Gender,drop_first = True)
```

```python
df.JobRole = df.JobRole.map({'Laboratory Technician':8,'Sales Executive':7,'Research Scientist':6,'Sales Representative':5,'Human Resources':4,'Manufacturing Director':3,'HealthCare Representative':2,'Manager':1,'Research Director':0})
```

```python
df.MartialStatus = df.MartialStatus.map({'Single':2,'Married':1,'Divorced':0})
```

```python
df.OverTime= df.OverTime.map({'yes':1,'No':0})
```
